Lea con OpenCV la imagen "alce.jpg" que está dentro de la carpeta "res". <img src="res/actividad1.png"/>

```python
###Escriba su código aquí###
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("res/alce.png", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

shape = img.shape

#Segundo cuadrante
img[0:int(shape[0]/2), int(shape[1]/2):, 0] = 0
img[0:int(shape[0]/2), int(shape[1]/2):, 1] = 0

#Tercer cuadrante
img[int(shape[0]/2):, :int(shape[1]/2), 0] = 0
img[int(shape[0]/2):, :int(shape[1]/2), 2] = 0

#Cuarto cuadrante
img[int(shape[0]/2):, int(shape[1]/2):, 1] = 0
img[int(shape[0]/2):, int(shape[1]/2):, 2] = 0

plt.imshow(img, cmap="gray")
```
